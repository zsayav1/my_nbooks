```python
# 1. Import the pandas module and the seaborn module.
import pandas as pd
import seaborn as sb
```

```python
# 2. Read avocado.csv into a Pandas DataFrame and display the first five rows.
avos = pd.read_csv("avocado.csv")
avos.head()
```

```python
# 3. Display basic information about the DataFrame and its columns. Note the Date column has the object type.
avos.info()
```

```python
# 4. Convert the Date column to the datetime type. Repeat #3.
avos["Date"] = pd.to_datetime(avos["Date"])
avos.info()
```

```python
""" 5. Filter the data with query() so it contains just the rows for 2015, for the conventional type, and for the region named Albany. 
Store only the Date, Total Bags, and Small Bags columns in a new DataFrame. Display the first 5 rows. """

# filtered data
avos.query('year == 2015' and 'type == "conventional"' and 'region == "Albany"').head()

# new data frame
avos2 = avos[['Date','Total Bags','Small Bags']]
avos2.head()
```

```python
""" 6. Melt the data in the Total Bags and Small Bags columns, but not the values in the Date column, naming the column that contains the type of bag 
as Bags and the column that contains the number of bags as Count. Display the melted DF. """

#Order is important in value_vars, as the output below will be different if switched.
avos2_long = avos2.melt(id_vars = "Date", value_vars = ['Total Bags', 'Small Bags'],
                        var_name = "Bags", value_name = "Count")
avos2_long 
```

```python
# 7. Plot the melted data with a Seaborn line plot, using Date as x-axis, Count as y-axis, and Bags for the hue. 
sb.relplot(data=avos2_long, kind = 'line', x = 'Date', y = 'Count', hue = 'Bags')
```

```python
""" 8. Bin the data in the Total Volume column into four quantiles labeled â€˜Poorâ€™, â€˜Modestâ€™, â€˜Goodâ€™, and â€˜Excellentâ€™, 
and store the bin labels in a new column named Sales Volume. Display the first five rows of the modified DF. """

avos['Sales Volume'] = pd.qcut(avos['Total Volume'], q=4, labels = ['Poor','Modest','Good','Excellent'])
avos.head()
```

```python
# 9. Plot the binned data by year with sns.countplot().
sb.countplot(data=avos, x = "year", hue = "Sales Volume")
```

```python
# 10. Melt the three bag size columns specifying var_name = Bags and value_name = Count. Display the melted DF.
avos3_long = avos.melt(id_vars = "year", value_vars = ['Small Bags', 'Large Bags', 'XLarge Bags'],
                        var_name = "Bags", value_name = "Count")
avos3_long
```

```python
# 11. Plot the bags by year using sns.relpot() with kind = line, x = year, y = Count and hue = Bags
sb.relplot(data=avos3_long, kind = "line", x = "year", y = "Count", hue = "Bags")
```
