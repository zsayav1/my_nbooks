## Fix spelling and capitalization problems in the data

```python
cars['brand'] = cars.apply(lambda x: x.CarName.split(' ')[0], axis=1)
cars['name'] = cars.apply(lambda x: ' '.join(x.CarName.split(' ')[1:]), axis=1)
```

```python
# 9. Display each unique value in the brand column.
cars.brand.unique()
```

```python
# 10. Use the replace() method to fix misspellings and inconsistent capitalization
cars.CarName.replace({'maxda':'mazda', 'Nissan':'nissan', 'porcshce':'porsche','toyouta':'toyota', 'vw':'volkswagen', 
                      'vokswagen':'volkswagen',}, inplace=True)
```

```python
# 11. Store the corrected data in the CarName column
cars.CarName = cars.brand + ' ' + cars.name
```

## Rename and drop columns

```python
# 13 Use the rename() method to rename CarName column to brandname and car_ID to carid
cars.rename(columns={'CarName':'brandname', 'car_ID':'carid'}, inplace=True)
```
